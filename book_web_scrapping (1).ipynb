Afrive project is about an e-commerce webisite that has only african books in its repository, the aim of this part of the project is to craete a recommender system that returns similar books to users on the website.
The data source to create this system is gotten from this [website](https://www.ascleiden.nl/content/webdossiers/africas-100-best-books-20th-century) and made to suit the features needed to run the model.

```python
#importing the necessary libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
# df= pd.read_html("/content/wikibook1.html")
url="https://www.ascleiden.nl/content/webdossiers/africas-100-best-books-20th-century"
response= requests.get(url)
response
```

```python
#reading the html text using beautiful soup
soup=BeautifulSoup(response.text)
print(soup.prettify())
```

```python
table = soup.select('table')[0]
tags= table.find_all('a')
```

```python
 links= [l.get("href") for l in tags]
 links
```
